```python
pets = pd.read_excel(path_pets)
users = pd.read_excel(path_users)
database = users.join(pets)

lista = [n for n in range(1, len(database) + 1)]
database['idUser'] = lista
database = database.set_index('idUser')
```

```python
def distancia_entre_vetores(a, b):
    return np.linalg.norm(a - b)
```

```python
def retorna_vetor_usuario(user_id):
    data = database
    vetor = []
    user = data.query('idUser == %d' % user_id)
    for value in user.columns[:7]: ## percorre as 7 primeiras colunas
        vetor.append(user[value].values[0])
    return np.array(vetor)
```

```python
def retorna_vetor_pet(user_id):
    data = database
    vetor = []
    user = data.query('idUser == %d' % user_id)
    for value in user.columns[7:]: ## percorre as 7 primeiras colunas
        vetor.append(user[value].values[0])
    return np.array(vetor)
```

```python
def distancia_entre_usuarios(user_id1, user_id2):
    distancia = distancia_entre_vetores(user_id1, user_id2)
    return [user_id1, user_id2, distancia]
```

```python
def distancia_entre_pets(user_id1, user_id2):
    distancia = distancia_entre_vetores(retorna_vetor_pet(user_id1), retorna_vetor_pet(user_id2))
    return [user_id1, user_id2, distancia]
```

```python
def distancia_de_todos_user(user_id, numero_de_usuarios_a_analisar = None):
    data = database
    todos_os_usuarios = data.index.unique()
    if numero_de_usuarios_a_analisar:
        todos_os_usuarios = todos_os_usuarios[:numero_de_usuarios_a_analisar]
    distancias = []
    for usuario in todos_os_usuarios:
        distancias.append([user_id, usuario, 
                           distancia_entre_usuarios(user_id, retorna_vetor_usuario(usuario))[2]])
    distancias = pd.DataFrame(distancias, columns = ['voce', 'outra_pessoa', 'distancia'])
    return(distancias)
```

```python
def distancia_de_todos_pet(user_id, numero_de_usuarios_a_analisar = None):
    data = database
    todos_os_usuarios = data.index.unique()
    if numero_de_usuarios_a_analisar:
        data = data[:numero_de_usuarios_a_analisar]
    distancias = []
    for usuario in todos_os_usuarios:
        distancias.append([user_id, usuario, 
                           distancia_entre_pets(user_id, usuario)[2]])
    distancias = pd.DataFrame(distancias, columns = ['seu pet', 'outro_pet', 'distancia'])
    return(distancias)
```

```python
def knn_user(user_id, k_mais_proximos=10, numero_de_usuarios_a_analisar = None):
    distancias = distancia_de_todos_user(user_id, numero_de_usuarios_a_analisar = numero_de_usuarios_a_analisar)
    distancias = distancias.sort_values("distancia")
    distancias = distancias.set_index("outra_pessoa")
    return distancias.head(k_mais_proximos)
```

```python
def knn_pet(user_id, k_mais_proximos=2, numero_de_usuarios_a_analisar = None):
    distancias = distancia_de_todos_pet(user_id, numero_de_usuarios_a_analisar = numero_de_usuarios_a_analisar)
    distancias = distancias.sort_values("distancia")
    distancias = distancias.set_index("outro_pet")
    distancias = distancias.drop(user_id)
    return distancias.head(k_mais_proximos)
```

```python
def load_data():
    data = pd.read_excel(path_database)
    data = data.drop('1 - De 0 a 10, o que você acha da ideia de utilizar inteligência artificial para recomendar pets? ', axis=1, errors='ignore')
    data = data.drop('2 - De 0 a 10, quão confiante você está de que uma IA pode fazer boas recomendações de pets com base nas suas características? ', axis=1, errors='ignore')
    data = data.drop('3 - De 0 a 10, quanto você acha que a utilização dessa ferramenta pode auxiliar as ONGs com a Adoção Responsável?', axis=1, errors='ignore')
    return data
```

```python
def caracteristicas_usuario(user_id):
    data = load_data()
    usuario = [data.values[user_id - 1][:7]]
    usuario = pd.DataFrame(usuario, columns=['Idade', 'Teve outro pet', 'Moradia', 'Crianças até 12 anos',
                                     'Recursos financeiros', 'Estilo de vida', 'Tempo em casa'])
    return usuario
```

```python
def caracteristicas_pets(user_id):
    data = load_data()
    pet = [data.values[user - 1][7:16] for user in user_id]
    pet = pd.DataFrame(pet, columns=['Idade', 'Nivel energia', 'Tamanho', 'Necessidade de estimulo',
                                     'Temperamento', 'Comportamento com crianca', 'Comportamento com animais',
                                     'Necessidades especiais', 'Solta muito pelo'])
    pet['idUser'] = user_id
    pet = pet.set_index('idUser')
    return pet
```

```python
def return_validation_array(users):
    D = [0 for n in users[0]]
    for user in users:
        for n in range(len(user)):
            D[n] += user[n]/len(users)
    
    return D
```

```python
def return_recommended_pets(user_id, k_users = 10, k_pets = 2):
    users_list = knn_user(user_id, k_mais_proximos=k_users).index
    recommended_pets = []
    for user in users_list:
        knn_pets = knn_pet(user, k_mais_proximos=k_pets).index.values
        for pet in knn_pets:
            recommended_pets.append(pet)
    try:
        recommended_pets.remove(user_id)
        return recommended_pets
    except:
        return recommended_pets
```

```python
def return_recommended_pets_table(recommended_pets):
    recommended_pets_arrays = [retorna_vetor_pet(pet) for pet in recommended_pets]
    validation_array = return_validation_array(recommended_pets_arrays)

    distances = [[]]
    for pet in recommended_pets:
        distances.append([pet, distancia_entre_vetores(validation_array, retorna_vetor_pet(pet))])

    recommended_pets_distances = pd.DataFrame(distances, columns=['pet_recomendado', 'distancia_pet_recomendado'])
    recommended_pets_distances = recommended_pets_distances.set_index('pet_recomendado')
    recommended_pets_table = recommended_pets_distances.join(caracteristicas_pets(recommended_pets)).dropna()
    recommended_pets_table = recommended_pets_table.sort_values('distancia_pet_recomendado')
    return recommended_pets_table
```
